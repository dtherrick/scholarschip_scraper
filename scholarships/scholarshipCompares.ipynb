```python
import warnings

warnings.filterwarnings('ignore')

import pandas as pd
from pathlib import Path

from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
datadir = Path("~/Data/scholarships").expanduser()

merged_scholarships = datadir / 'mergedScholarshipInfo.csv'
nathan_scholarships = datadir / 'nathanList.csv'

outfile = datadir / 'matchedScholarships.csv'
```

```python
df_allScholarships = pd.read_csv(merged_scholarships)
```

```python
df_nathanScholarships = pd.read_csv(nathan_scholarships, header=None, names=['Scholarship'])
```

```python
df_allScholarships.columns.to_list()
```

```python
def checker(wrong_options,correct_options):
    names_array=[]
    ratio_array=[]    
    for wrong_option in wrong_options:
        if wrong_option in correct_options:
            names_array.append(wrong_option)
            ratio_array.append('100')
        else:   
            x=process.extractOne(wrong_option,correct_options,scorer=fuzz.token_set_ratio)
            names_array.append(x[0])
            ratio_array.append(x[1])
    return names_array,ratio_array
```

```python
source_list = df_nathanScholarships['Scholarship'].fillna('####').to_list()
to_match_list_organizations = df_allScholarships['Organization_left'].fillna('####').to_list()
to_match_list_awards = df_allScholarships['Award Name'].fillna('####').to_list()
```

```python
name_match_organizations,ratio_match_organizations = checker(source_list,to_match_list_organizations)

df_organization_match = pd.DataFrame()
df_organization_match['initial_names']=pd.Series(source_list)
df_organization_match['correct_names']=pd.Series(name_match_organizations)
df_organization_match['correct_ratio']=pd.Series(ratio_match_organizations)
```

```python
name_match_awards,ratio_match_awards = checker(source_list,to_match_list_awards)

df_awards_match = pd.DataFrame()
df_awards_match['initial_names']=pd.Series(source_list)
df_awards_match['correct_names']=pd.Series(name_match_awards)
df_awards_match['correct_ratio']=pd.Series(ratio_match_awards)
```
